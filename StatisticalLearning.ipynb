```python
import numpy as np
from numpy.random import rand , randn

import matplotlib.pyplot as plt

def generate_data(beta , sig, n):  # generate data with the requested variance given by sig**2
   u = np.random.rand(n, 1)
   y = (u ** np.arange(0, 4)) @ beta + sig * np.random.randn(n, 1) # we use @ for matrix multiplication
   return u, y

np.random.seed(12)
beta = np.array([[10, -140, 400, -250]]).T
n = 100
sig = 10
u, y = generate_data(beta , sig, n)

# define the smooth line
xx = np.arange(np.min(u), np.max(u)+5e-3, 5e-3)
yy = np.polyval(np.flip(beta), xx)     # evaluate the polynomial

plt.plot(u, y, '.', markersize=8)
plt.plot(xx, yy, '--',linewidth=3)

# %%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
plt.xlabel(r'$u$')
plt.ylabel(r'$h^*(u)$')
plt.legend(['data points','true'])
plt.savefig('../figures/polydatpy.pdf',format='pdf')
plt.show()
# %%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
```

```python
import random
import numpy as np
import matplotlib.pyplot as plt


a = -2
b= 3
N = 1000

c = -1
d = 0.5
j=0
ar = np.zeros(N)
ar2=np.zeros(N)
for i in range (len(ar)):
    ar[i] = random.uniform(a,b)
    ar2[i] = random.uniform(c,d)
integral = 0.0
def f(x,y):
    return x*x+3*y*y+1
while j<N:
    integral += f(ar[j],ar2[j])
    j+=1
ans = (b-a)*(d-c)/float(N)*integral
print ("Approx to the integral by CMC: {}.".format(ans))
# Define the range for x and y
x = np.linspace(-2, 3, 1000)
y = np.linspace(-1, 0.5, 1000)

# Create a grid of x and y values
X, Y = np.meshgrid(x, y)

# Calculate the corresponding z values using the function x*2 + 3*y*2 + 1
Z = X*2 + 3*Y*2 + 1

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
ax.scatter(X, Y, Z, c='b', marker='o', label='Scatter')

# Create a 3D surface plot
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8, label='Surface')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Scatter and Surface Plot')

# Show theplot
plt.show()
```
